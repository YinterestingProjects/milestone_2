```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
```

```python
# read in consolidated data file
df = pd.read_csv('../data/Tara/consolidated_data_2016_2020_2.csv')
```

```python
df.sample(5)
```

```python
# drop irrelevant columns
drop_cols = ['Unnamed: 0']
df = df.drop(columns=drop_cols)
```

```python
# check for duplicate rows
duplicated_df = df[df.duplicated(keep='first')]
print(f'number of duplicated rows {duplicated_df.shape[0]:,} out of {df.shape[0]:,} total rows')

# drop duplicated rows
dedup_df = df.drop_duplicates(ignore_index=True)
print(f'rows remaining {dedup_df.shape[0]:,}')
```

```python
duplicated_df.head()
```

```python
# save duplicated df to data folder path 
path = '../data/Tara/'
output_name = 'duplicated.csv'
duplicated_df.to_csv(f'{path}{output_name}')
```

```python
def top10(df):
    """Given dataframe, generate top 10 unique values for non-numeric data"""
    columns = df.select_dtypes(include=['object', 'category']).columns
    for col in columns:
        print("Top 10 unique values of " + col)
        print(df[col].value_counts().reset_index(
        ).rename(columns={"index": col, col: "Count"}
                )[:min(10, len(df[col].value_counts()))])
        print(" ")
        
def categorical_eda(df, hue=None):
    """Given dataframe, generate EDA of categorical data"""
    print("\nTo check: \nUnique count of non-numeric data\n")
    print(df.select_dtypes(include=['object', 'category']).nunique())
    ##top5(df)
    # Plot count distribution of categorical data
    for col in df.select_dtypes(include=['object', 'category']).columns:
        
        fig = sns.catplot(x=col, kind="count", data=df, hue=hue)
        fig.set_xticklabels(rotation=90)
        plt.show()
    
```
