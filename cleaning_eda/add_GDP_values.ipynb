```python
import pandas as pd
import numpy as np
```

```python
p = '../data/'
input_name = 'cleaned.csv'
df = pd.read_csv(f'{p}{input_name}')
```

```python
df['ctry_org'] = df['ctry_org'].str.upper()
df['ctry_ie'] = df['ctry_ie'].str.upper()
df.head()
```

```python
def cleanup_nulls(df, column_names):
    names = column_names
    for name in names:
        df.loc[df[name].isnull(), name] = "NaN_"+name
    return df

cleanup_nulls(df, ['ctry_org','ctry_ie'])
```

```python
# bring in imf dataset
p = '../data/'
input_name = 'IMF_GDP_per_capita.csv'
imf = pd.read_csv(f'{p}{input_name}')

imf = imf[['Country','2020']]
imf.head()
```

```python
# bring in country code dataset
p = '../data/'
input_name = 'key_2016.csv'
ctry = pd.read_csv(f'{p}{input_name}', encoding="cp1252").drop([275,286,287,288,289,290,291])

ctry = ctry[ctry['field']=='country']
ctry.head()
```

```python
# merge df & ctry keys 
df_ctry = df.merge(ctry, how='left', left_on='ctry_ie', right_on='code')
df_ctry = df_ctry.rename(columns={'value_y':'ctry_ie_name'})
df_ctry.head()
```

```python
ctry_dict = {'G5':'United Kingdom','G2':'United Kingdom','G4':'United Kingdom','G1':'United Kingdom',
            'G6':'United Kingdom','G3':'United Kingdom','E1':'Germany','GB':'United Kingdom'}
for k, v in ctry_dict.items():
    if df_ctry['ctry_ie'].str.contains(k).any():
        index_nums = df_ctry[df_ctry['ctry_ie']==k].index
        for index_num in index_nums:
            df_ctry.at[index_num,'ctry_ie_name']=v
```

```python
# merge df_ctry with imf dataset
df_ctry_imf = df_ctry.merge(imf, how='left', left_on='ctry_ie_name', right_on='Country')
df_ctry_imf = df_ctry_imf.rename(columns={'2020':'2020GDP'}).drop(columns=['field','code','Country'])
df_ctry_imf
```

```python
cleanup_nulls(df_ctry_imf, ['ctry_ie_name'])
```

```python
# create a dataset = cleaned.csv + GDP values
p = '../data/'
input_name = 'cleaned_GDP.csv'
df_ctry_imf.to_csv(f'{p}{input_name}')
```
